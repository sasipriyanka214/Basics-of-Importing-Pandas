```python
# Import necessary libraries
import pandas as pd

# Read the CSV file into a Pandas DataFrame and ensure TransactionDate is parsed as datetime
df = pd.read_csv('retail_transactions.csv', parse_dates=['TransactionDate'])
```

```python
df
```

```python
# Display the first 15 rows of the DataFrame
print("First 15 rows of the DataFrame:")
print(df.head(15))

# Print summary information about the DataFrame
print("\nSummary of the DataFrame:")
print(df.info())

# Print descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# Check for missing values in each column and print the count per column
print("\nMissing Values Count Per Column:")
print(df.isnull().sum())

# Fill missing values in the Discount column with 0 (assuming no discount if missing)
df['Discount'] = df['Discount'].fillna(0)

# Convert Quantity, Price, and Discount to numeric types (if not already)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')
```

```python
# Create a new column NetPrice that applies the discount to the price
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)

# Create another new column TotalSale computed as Quantity * NetPrice
df['TotalSale'] = df['Quantity'] * df['NetPrice']
```

```python
df['NetPrice']
df['TotalSale']
```

```python
# Filter the DataFrame to include only transactions where TotalSale >= 20
filtered_df = df[df['TotalSale'] >= 20]
```

```python
filtered_df
```

```python
# Convert the PaymentType column to a categorical data type
filtered_df['PaymentType'] = filtered_df['PaymentType'].astype('category')
```

```python
filtered_df['PaymentType']
```

```python
# Group by ProductCategory and compute total Quantity sold and total TotalSale for each category
grouped_by_category = filtered_df.groupby('ProductCategory').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalSales=('TotalSale', 'sum')
).reset_index()

# Sort the grouped result in descending order based on TotalSales
grouped_by_category = grouped_by_category.sort_values(by='TotalSales', ascending=False)

# Identify and print the top 3 product categories with the highest total sales
print("\nTop 3 Product Categories by Total Sales:")
print(grouped_by_category.head(3))
```

```python
# Group the filtered data by StoreID and ProductCategory
multi_grouped = filtered_df.groupby(['StoreID', 'ProductCategory']).agg(
    AvgTotalSalePerTransaction=('TotalSale', 'mean')
).reset_index()
```

```python
multi_grouped
```

```python
# For each store, determine which product category has the highest average TotalSale
store_top_categories = multi_grouped.loc[multi_grouped.groupby('StoreID')['AvgTotalSalePerTransaction'].idxmax()]
store_top_categories = store_top_categories[['StoreID', 'ProductCategory', 'AvgTotalSalePerTransaction']]

# Rename columns for clarity
store_top_categories.rename(columns={'AvgTotalSalePerTransaction': 'MaxAvgTotalSale'}, inplace=True)

# Print the summary DataFrame
print("\nStore-Level Insights: Top Product Category by Average Total Sale")
print(store_top_categories)
```
